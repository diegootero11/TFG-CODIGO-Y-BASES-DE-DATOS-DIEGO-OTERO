```python

nulos_cash_before = df_MAs['Percentage_of_Cash'].isnull().sum()
nulos_stock_before = df_MAs['Percentage_of_Stock'].isnull().sum()

print(f"Valores nulos en Percentage_of_Cash antes de la imputaciÃ³n: {nulos_cash_before}")
print(f"Valores nulos en Percentage_of_Stock antes de la imputaciÃ³n: {nulos_stock_before}")


df_MAs.loc[df_MAs['Consideration_Structure'] == 'Stock Only', 'Percentage_of_Cash'] = 0
df_MAs.loc[df_MAs['Consideration_Structure'] == 'Stock Only', 'Percentage_of_Stock'] = 100

df_MAs.loc[df_MAs['Consideration_Structure'] == 'Cash Only', 'Percentage_of_Cash'] = 100
df_MAs.loc[df_MAs['Consideration_Structure'] == 'Cash Only', 'Percentage_of_Stock'] = 0


nulos_cash_after = df_MAs['Percentage_of_Cash'].isnull().sum()
nulos_stock_after = df_MAs['Percentage_of_Stock'].isnull().sum()

print(f"Valores nulos en Percentage_of_Cash despuÃ©s de la imputaciÃ³n: {nulos_cash_after}")
print(f"Valores nulos en Percentage_of_Stock despuÃ©s de la imputaciÃ³n: {nulos_stock_after}")
```

```python

categories_to_process = [
    'Cash and Stock Combination',
    'Choice between\nCash or Stock or\nCombination of Both'
]

for category in categories_to_process:

    category_filter = df_MAs['Consideration_Structure'] == category


    df_MAs.loc[
        category_filter & df_MAs['Percentage_of_Cash'].notnull() & df_MAs['Percentage_of_Stock'].isnull(),
        'Percentage_of_Stock'
    ] = 100 - df_MAs.loc[
        category_filter & df_MAs['Percentage_of_Cash'].notnull() & df_MAs['Percentage_of_Stock'].isnull(),
        'Percentage_of_Cash'
    ]


    df_MAs.loc[
        category_filter & df_MAs['Percentage_of_Stock'].notnull() & df_MAs['Percentage_of_Cash'].isnull(),
        'Percentage_of_Cash'
    ] = 100 - df_MAs.loc[
        category_filter & df_MAs['Percentage_of_Stock'].notnull() & df_MAs['Percentage_of_Cash'].isnull(),
        'Percentage_of_Stock'
    ]


nulos_cash_after = df_MAs['Percentage_of_Cash'].isnull().sum()
nulos_stock_after = df_MAs['Percentage_of_Stock'].isnull().sum()

print(f"Valores nulos en Percentage_of_Cash despuÃ©s del tratamiento: {nulos_cash_after}")
print(f"Valores nulos en Percentage_of_Stock despuÃ©s del tratamiento: {nulos_stock_after}")
```

```python

def get_gdp_for_row(row, df_gdp):
    country = row['Target_Nation']
    date_announced = row['Date_Announced']
    date_effective = row['Date_Effective_or_Withdraw']


    quarter_announced = (date_announced.month - 1) // 3 + 1 if pd.notna(date_announced) else None
    quarter_effective = (date_effective.month - 1) // 3 + 1 if pd.notna(date_effective) else None


    gdp_announce = None
    gdp_effective = None

    if pd.notna(quarter_announced):
        gdp_announce = df_gdp.loc[(df_gdp['DATE'].dt.year == date_announced.year) &
                                   (df_gdp['DATE'].dt.quarter == quarter_announced), country].values
    if pd.notna(quarter_effective):
        gdp_effective = df_gdp.loc[(df_gdp['DATE'].dt.year == date_effective.year) &
                                    (df_gdp['DATE'].dt.quarter == quarter_effective), country].values


    gdp_announce = gdp_announce[0] if len(gdp_announce) > 0 else None
    gdp_effective = gdp_effective[0] if len(gdp_effective) > 0 else None

    return pd.Series([gdp_announce, gdp_effective])


df_MAs[['GDP_Announced_Target', 'GDP_Effective_Target']] = df_MAs.apply(lambda row: get_gdp_for_row(row, df_gdp_growth), axis=1)


print(df_MAs[['Target_Nation', 'Date_Announced', 'Date_Effective_or_Withdraw', 'GDP_Announced_Target', 'GDP_Effective_Target']].head())
```

```python

def get_macro_data_for_date(row, df_macro, variable_name):

    country = row['Target_Nation']
    date_announced = row['Date_Announced']
    date_effective = row['Date_Effective_or_Withdraw']


    data_announced = df_macro.loc[(df_macro['DATE'].dt.month == date_announced.month) &
                                  (df_macro['DATE'].dt.year == date_announced.year), country].values

    data_effective = df_macro.loc[(df_macro['DATE'].dt.month == date_effective.month) &
                                  (df_macro['DATE'].dt.year == date_effective.year), country].values


    if len(data_announced) == 0:
        data_announced = None
    else:
        data_announced = data_announced[0]

    if len(data_effective) == 0:
        data_effective = None
    else:
        data_effective = data_effective[0]

    return pd.Series([data_announced, data_effective])


df_MAs[['CPI_Announced_Target', 'CPI_Effective_Target']] = df_MAs.apply(
    lambda row: get_macro_data_for_date(row, df_CPI, 'CPI'), axis=1)



df_MAs[['Interest_Rate_Announced_Target', 'Interest_Rate_Effective_Target']] = df_MAs.apply(
    lambda row: get_macro_data_for_date(row, df_interest_rates, 'Interest Rates'), axis=1)


df_MAs[['M2_Announced_Target', 'M2_Effective_Target']] = df_MAs.apply(
    lambda row: get_macro_data_for_date(row, df_m2, 'M2'), axis=1)


print(df_MAs[['Target_Nation', 'Date_Announced', 'Date_Effective_or_Withdraw',
              'CPI_Announced_Target', 'CPI_Effective_Target',
              'GDP_Announced_Target', 'GDP_Effective_Target',
              'Interest_Rate_Announced_Target', 'Interest_Rate_Effective_Target',
              'M2_Announced_Target', 'M2_Effective_Target']].head())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency


categorical_columns = df_MAs.select_dtypes(include='category').columns


def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix, correction=False)[0]
    n = confusion_matrix.sum().sum()
    k_min = min(confusion_matrix.shape) - 1
    return np.sqrt(chi2 / (n * k_min)) if k_min > 0 else np.nan


cramers_matrix = pd.DataFrame(
    np.zeros((len(categorical_columns), len(categorical_columns))),
    index=categorical_columns,
    columns=categorical_columns
)


for col1 in categorical_columns:
    for col2 in categorical_columns:
        if col1 == col2:
            cramers_matrix.loc[col1, col2] = 1.0
        else:
            cramers_matrix.loc[col1, col2] = cramers_v(df_MAs[col1], df_MAs[col2])


plt.figure(figsize=(12, 10))
sns.heatmap(cramers_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=0, vmax=1)
plt.title("Matriz de CramÃ©r's V entre variables categÃ³ricas", fontsize=16)
plt.tight_layout()
plt.show()
```
