```python

df_ipos['Dates_Issue_Date'] = pd.to_datetime(df_ipos['Dates_Issue_Date'], errors='coerce')



def get_quarter(date):
    return (date.month - 1) // 3 + 1 if pd.notna(date) else None


df_ipos.loc[:, 'Quarter_Issue'] = df_ipos['Dates_Issue_Date'].apply(get_quarter)



print(df_ipos[['Dates_Issue_Date', 'Quarter_Issue']].head())
```

```python

def get_gdp_for_row(row, df_gdp):
    country = row['IssuerBorrower_Nation']
    date_announced = row['Dates_Issue_Date']



    quarter_announced = (date_announced.month - 1) // 3 + 1 if pd.notna(date_announced) else None



    gdp_announce = None


    if pd.notna(quarter_announced):
        gdp_announce = df_gdp.loc[(df_gdp['DATE'].dt.year == date_announced.year) &
                                   (df_gdp['DATE'].dt.quarter == quarter_announced), country].values


    gdp_announce = gdp_announce[0] if len(gdp_announce) > 0 else None


    return pd.Series([gdp_announce])


df_ipos[['GDP_Announced']] = df_ipos.apply(lambda row: get_gdp_for_row(row, df_gdp_growth), axis=1)


print(df_ipos[['IssuerBorrower_Nation', 'Dates_Issue_Date', 'GDP_Announced']].head())
```

```python

def get_macro_data_for_date(row, df_macro, variable_name):

    country = row['IssuerBorrower_Nation']
    date_announced = row['Dates_Issue_Date']



    data_announced = df_macro.loc[(df_macro['DATE'].dt.month == date_announced.month) &
                                  (df_macro['DATE'].dt.year == date_announced.year), country].values



    if len(data_announced) == 0:
        data_announced = None
    else:
        data_announced = data_announced[0]



    return pd.Series([data_announced])


df_ipos[['CPI_Announced']] = df_ipos.apply(
    lambda row: get_macro_data_for_date(row, df_CPI, 'CPI'), axis=1)



df_ipos[['Interest_Rate_Announced']] = df_ipos.apply(
    lambda row: get_macro_data_for_date(row, df_interest_rates, 'Interest Rates'), axis=1)


df_ipos[['M2_Announced']] = df_ipos.apply(
    lambda row: get_macro_data_for_date(row, df_m2, 'M2'), axis=1)


print(df_ipos[['IssuerBorrower_Nation', 'Dates_Issue_Date',
              'CPI_Announced',
              'GDP_Announced',
              'Interest_Rate_Announced',
              'M2_Announced']].head())
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt


cat_vars = [
    'IssuerBorrower_Nation',
    'IssuerBorrower_Stock_Exchange_Name',
    'Offering_Technique',
    'Security_Type_All_Markets',
    'IssuerBorrower_TRBC_Industry'
]


def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = stats.chi2_contingency(confusion_matrix, correction=False)[0]
    n = confusion_matrix.sum().sum()
    k_min = min(confusion_matrix.shape) - 1
    return np.sqrt(chi2 / (n * k_min)) if k_min > 0 else np.nan


cramers_matrix = pd.DataFrame(np.zeros((len(cat_vars), len(cat_vars))), index=cat_vars, columns=cat_vars)


for col1 in cat_vars:
    for col2 in cat_vars:
        if col1 == col2:
            cramers_matrix.loc[col1, col2] = 1.0
        else:
            cramers_matrix.loc[col1, col2] = cramers_v(df_ipos[col1], df_ipos[col2])

plt.figure(figsize=(10, 8))
sns.heatmap(cramers_matrix, annot=True, cmap="coolwarm", fmt=".2f", vmin=0, vmax=1)
plt.title("Matriz de Cramér's V entre variables categóricas", fontsize=16)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf


target = 'Proceeds_Amount_Incl_Overallotment_Sold_All_Markets_USD_Millions_ΣCom'


cat_vars = [
    'IssuerBorrower_Nation',
    'IssuerBorrower_Stock_Exchange_Name',
    'Offering_Technique',
    'Security_Type_All_Markets',
    'IssuerBorrower_TRBC_Industry'
]


def eta_squared(df, cat_var, target):
    model = smf.ols(f'{target} ~ C({cat_var})', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    ss_between = anova_table['sum_sq'][0]
    ss_total = sum((df[target] - df[target].mean())**2)
    return round(ss_between / ss_total, 4)


resultados_eta = [(var, eta_squared(df_ipos, var, target)) for var in cat_vars]
df_eta = pd.DataFrame(resultados_eta, columns=['Variable Categórica', 'η² (varianza explicada)'])


from IPython.display import display
display(df_eta)
```
